```r
library(ggplot2)
library(dplyr) # needed for filter and other things
```

```r
# importing csv files 
dc = read.csv("./data/comics/dc.csv", stringsAsFactors = TRUE)
marvel = read.csv("./data/comics/marvel.csv", stringsAsFactors = TRUE)
```

```r
# adding new column about the comic the owns the character
dc['publisher'] <- 'DC'
marvel['publisher'] <- 'MARVEL'

# making column names lowercase
colnames(dc) <- tolower(colnames(dc))
colnames(marvel) <- tolower(colnames(marvel))

# checking column names to ensure they are same
colnames(dc)
colnames(marvel)

# merging the data to create one dataset
comics <- rbind(dc, marvel)

# updating the factor levels to make them compact
levels(comics$align) <- list('Bad' = 'Bad Characters','Good' = 'Good Characters','Neutral' = 'Neutral Characters','Reformed' = 'Reformed Criminals')
levels(comics$id) <- list('Unknown' = 'Identity Unknown','Public' = 'Public Identity','Secret' = 'Secret Identity','Authorities Only' = 'Known to Authorities Identity', 'None' = 'No Dual Identity')
levels(comics$sex) <- list('Female' = 'Female Characters','Genderless' = 'Genderless Characters','Male' = 'Male Characters','Transgender' = 'Transgender Characters', 'Agender' = 'Agender Characters', 'Genderfluid'='Genderfluid Characters')
levels(comics$gsm) <- list('Bisexual'='Bisexual Characters', 'Homosexual' = 'Homosexual Characters', 'Genderfluid' = 'Genderfluid Characters', 'Pansexual' = 'Pansexual Characters', 'Transgender' = 'Transgender Characters', 'Transvestites' = 'Transvestites')
levels(comics$alive) <- list('Deceased' = 'Deceased Characters', 'Living' = 'Living Characters')

# making column names intitutive 
colnames(comics) <- c('page','name','url','identity','align','eye','hair','sex','sexuality','alive','appearances','first.appearance','year','publisher')

```

```r
str(comics)
head(comics)
```

```r
comics_filtered <- comics %>% filter(align != "Reformed") %>% droplevels()
```

```r
par(mfrow=c(1,2))
ggplot(dc, aes(x=align, fill =sex)) + geom_bar(position = position_dodge())
ggplot(dc, aes(x=sex, fill =align)) + geom_bar(position = position_dodge())
```

```r
options(scipen = 999, digits = 3)
table(comics$id, comics$align)
prop.table(table(comics$id, comics$align))
sum(prop.table(table(comics$id, comics$align)))
```

```r
print("#conditional proportion along the row")
prop.table(table(comics$id, comics$align),1) #conditional proportion along the row
print("#conditional proportion along the column")
prop.table(table(comics$id, comics$align),2) #conditional proportion along the column
```

```r
#shows the proportion of character type within each type of identity
# ie, within secret identity, how many bad how many good etc
ggplot(comics, aes(x=identity, fill=align)) +
geom_bar() + 
ylab("proportion")
```

```r
#shows the proportion of identity type within each type of character type
# ie, within bad character type, how many secret identity how many public identity etc
ggplot(comics, aes(x=align, fill=identity)) +
geom_bar(position = "fill") + 
ylab("proportion")

# this seem paradoxical to above graph as i just exchanged the column names but it is so becasue of number of cases in each.
```

```r
tab <- table(comics$align, comics$sex)
options(scipen = 999, digits = 3) # Print fewer digits
prop.table(tab)     # Joint proportions
prop.table(tab, 2)  # Conditional on columns
```

```r
# Faceted bar, instead of combining values in same graph, splitted in different graphs
ggplot(comics, aes(x=identity, fill=align)) +
geom_bar() + 
facet_wrap(~align)
```
